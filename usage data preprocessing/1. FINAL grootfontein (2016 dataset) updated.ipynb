```python
#create a series in list comprehension and join them together using concat, replacing missing values, convert to intergers and last convert index to column date 
L = [pd.Series(v, index=range(int(s), int(e)+1)) for s,e,v in 
     zip(new_set['start year'], new_set['end year'], new_set['Registered Volume'],)]
df1 = (pd.concat(L, axis=1, keys=new_set['Register No.'])
         .fillna(0)
         .astype(int)
         .rename_axis('date')
         .reset_index())
```
