```python
df = pd.read_excel('/Users/yolanda/Documents/Priority compartments/Grootfontein/gwls/fixed workbook with all boreholes on 1 sheet.xlsx') 
```

```python
#using the following code to subset your dataframe by BH_ID and save each resultant subsetted dataframe to the dictionary df_dict:

df_dict = {}

for data in df['BH_ID'].unique():
    data_df = df[df['BH_ID'] == data]
    df_dict[data] = data_df
```

```python
#Once i have this dictionary, am using pd.ExcelWriter and to_excel to write my dataframes into separate sheets of an excel file:
def save_xlsx(df_dict, path):
    """
    Save a dictionary of dataframes to an excel file, with each dataframe as a seperate page
    """

    with pd.ExcelWriter(path) as writer:
        for key in df_dict:
            df_dict[key].to_excel(writer, key, index=False)

    writer.save()
```

```python
# Calling the function will give the desired result:
save_xlsx(df_dict, '/Users/yolanda/Documents/Priority compartments/Grootfontein/gwls/workbook with individual borehole sheets.xlsx')
```
