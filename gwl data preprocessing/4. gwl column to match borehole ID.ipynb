```python
from pathlib import Path

folder = Path('/Users/yolanda/Documents/Priority compartments/Grootfontein/gwls/final individual boreholes/')

xlsx_only_files = list(folder.rglob('*.xlsx'))
```

```python
dicts = [pd.read_excel(df, usecols=['year_month','GWL'], sheet_name = None) 
         for df in xlsx_only_files]
filenames = [f.stem for f in xlsx_only_files]
```

```python
new_dict = {}
for file_idx in range(len(dicts)):
    dicts[file_idx]['Sheet1'].rename(columns = {'GWL': filenames[file_idx]}, inplace=True)
    new_dict.update({filenames[file_idx]: dicts[file_idx]['Sheet1']})
    dicts[file_idx]['Sheet1'].to_excel('/Users/yolanda/Documents/Priority compartments/Grootfontein/gwls/final individual boreholes/untitled folder/' + filenames[file_idx] + '.xlsx')
```
