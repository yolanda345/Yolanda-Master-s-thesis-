```python
L1 = []
for s1, df1 in pd.read_excel('/Users/yolanda/Downloads/1Grootfontein_sheets.xlsx', sheet_name=None).items():
    d1 = {'sheet name': s1, 
          'latitude': df1['Latitude'].min(), 
          'longitude': df1['Longitude'].min(), 
         'start date': pd.to_datetime(df1['year_month']).min(),
         'end date': pd.to_datetime(df1['year_month']).max(),
         'missing values':df1['GWL'].isna().sum()}
    L1.append(d1)

df1 = pd.DataFrame(L1)
```

```python
L2 = []
for s2, df2 in pd.read_excel('/Users/yolanda/Downloads/2Grootfontein_sheets.xlsx', sheet_name=None).items():
    d2 = {'sheet name': s2, 
          'latitude': df2['Latitude'].min(), 
          'longitude': df2['Longitude'].max(), 
         'start date': pd.to_datetime(df2['year_month']).min(),
         'end date': pd.to_datetime(df2['year_month']).max(),
         'missing values':df2['GWL'].isna().sum()}
    L2.append(d2)

df2 = pd.DataFrame(L2)
```

```python
total_years1 = df1['end date'].apply(lambda x: x.year) - df1['start date'].apply(lambda x: x.year)
```

```python
total_years2 = df2['end date'].apply(lambda x: x.year) - df2['start date'].apply(lambda x: x.year)
```

```python
df1['total Years Running'] = total_years1
```

```python
df2['total Years Running'] = total_years2
```

```python
df1
```

```python
df2
```

```python
months1 = df1['total Years Running']*12
```

```python
months2 = df2['total Years Running']*12
```

```python
percentage1 = df1['missing values']/months1*100
```

```python
percentage2 = df2['missing values']/months2*100
```

```python
df1['missing percentage'] = percentage1
```

```python
df2['missing percentage'] = percentage2
```

```python
df1
```

```python
df2
```

```python
df1['range'] = pd.cut(df1['missing percentage'],bins=[0,25,50,75,100], labels=[1,2,3,4])
```

```python
df2['range'] = pd.cut(df2['missing percentage'],bins=[0,25,50,75,100], labels=[1,2,3,4])
```

```python
frames = [df1, df2]

result = pd.concat(frames)
```
